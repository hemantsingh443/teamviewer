```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_images_from_imgset(band_dir, imgset_name, title):
    imgset_path = os.path.join(band_dir, imgset_name)
    hr_path = os.path.join(imgset_path, 'HR.png')
    lr000_path = os.path.join(imgset_path, 'LR000.png')
    lr001_path = os.path.join(imgset_path, 'LR001.png')

    images_to_display = []
    if os.path.exists(hr_path):
        images_to_display.append(mpimg.imread(hr_path))
    if os.path.exists(lr000_path):
        images_to_display.append(mpimg.imread(lr000_path))
    if os.path.exists(lr001_path):
        images_to_display.append(mpimg.imread(lr001_path))

    if not images_to_display:
        print(f"No images found for {imgset_name} in {band_dir}")
        return

    fig, axes = plt.subplots(1, len(images_to_display), figsize=(15, 5))
    if len(images_to_display) == 1:
        axes = [axes]

    fig.suptitle(title, y=1.02, fontsize=16)

    for i, img in enumerate(images_to_display):
        axes[i].imshow(img)
        axes[i].axis('off')
        if i == 0:
            axes[i].set_title('HR')
        elif i == 1:
            axes[i].set_title('LR000')
        elif i == 2:
            axes[i].set_title('LR001')


    plt.tight_layout()
    plt.show()

red_band_dir = 'data/RED'
display_images_from_imgset(red_band_dir, 'img1', 'RED Band - Sample 1')


nir_band_dir = 'data/NIR'
display_images_from_imgset(nir_band_dir, 'img1', 'NIR Band - Sample 1')
```
