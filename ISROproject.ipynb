```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the images
lr0_img = mpimg.imread('/data/img1/LR000.png')
lr1_img = mpimg.imread('/data/img1/LR001.png')
hr_img = mpimg.imread('//data/img1/HR.png')

# Create a figure with 2 rows and 2 columns
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Display the LR images in the first row
axes[0, 0].imshow(lr0_img)
axes[0, 0].set_title('LR0')
axes[0, 0].axis('off') # Hide axes

axes[0, 1].imshow(lr1_img)
axes[0, 1].set_title('LR1')
axes[0, 1].axis('off') # Hide axes

# Display the HR image in the second row, spanning both columns
axes[1, 0].imshow(hr_img)
axes[1, 0].set_title('HR')
axes[1, 0].axis('off') # Hide axes

# Remove the empty subplot in the second row, second column
fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```
